# 気象データから1時間後の降雨を予測する

過去～現時点までのデータから１時間後の降雨の有無を予測する

***

## １．データの概要を把握する

### CSVファイルからデータを読み込む

```python
#ライブラリーを読み込む
import pandas as pd
pd.set_option('precision', 2)

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import datetime

# データの読込と表示
df_orig = pd.read_csv(
    "17_preprocessed.csv",
    sep="," ,
    skiprows = 1,
    header= 1,
    encoding = "Shift_JISx0213").set_index("年月日時")
df_orig.head()            # 統計量の表示
```

***

### 統計データを表示

```python
df_orig.describe()
```

***

### データを標準化する

```python
df =  (df_orig - df_orig.mean()) / (df_orig.std())
df.head()
```

```python
df.describe()
```

### 相関係数を表示

```python
df_corr = df.corr()
df_corr
```

### 相関係数の大きい要素のみ表示

```python
df_corr.applymap(lambda x: x if abs(x) > 0.3 else "---")
```

### 降水量と他データとの相関

```python
pd.reset_option("display.precision")
pd.DataFrame(df_corr["降水量(mm)"])
```

## → <font color="red">残念なから降水量と相関の強いデータはなし</font>

***

## 方向づけ  
* 相関係数からでは降水量の予測に適した要素を見つけられなかった。
* 決定木から降水量に関連性の強い要素を見つける
* 関連性の強い要素から降水量を推定する<br>

## 学習設定
* 決定木の階層（＝if文のネストの深さ）は３とする
* 1年分のデータの内、1月～11月までのデータをトレーニングデータとして、12月をテストデータとする

```python
MAX_NEST_DEPTH = 3
TEST_DATA_LENGTH = 24 * 30  # 1か月分をテストデータとする
```

### データ準備

```python
src = np.array(df.drop('降水量(mm)', axis=1))
src = np.delete(src, -1, axis = 0)  # 最終のデータは予測には使用しない（できない）ため削除
```

```python
target = np.array(df['降水量(mm)'])
target = np.delete(target, 0, axis = 0)  # 先頭のデータは予測できないため削除
#target = np.round(target) # DecisionTreeClassifierは整数である必要あり。
target = [(True if t > min(target) else False) for t in target] # 雨が少しでも降れば「降雨あり」
```

```python
plt.subplots(figsize=(13, 5))
index = [datetime.datetime.strptime(i, '%Y/%m/%d %H:%M') for i in df.index]
plt.plot(index, df['降水量(mm)'].values)
plt.show()
```

***
## トレーニング開始

### 決定木を実行

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier(max_depth=MAX_NEST_DEPTH)
#clf = tree.DecisionTreeRegressor(max_depth=3)  # 降水量が小数でもOKだが..これは一体何？
clf = clf.fit(src[:-TEST_DATA_LENGTH], target[:-TEST_DATA_LENGTH])
```

```python
# 作成した決定木を可視化 (pydotplus パッケージを利用)
import pydotplus
from sklearn.externals.six import StringIO
dot_data = StringIO()
tree.export_graphviz(clf, out_file=dot_data)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
```

```python
from IPython.display import Image
Image(graph.create_png())
```

```python
for (i, n) in enumerate(df.drop('降水量(mm)', axis=1).columns):
    print(i, n)
```

### トレーニングした結果

```python
predicted = clf.predict(src[:-TEST_DATA_LENGTH])
print(len(target[:-TEST_DATA_LENGTH]), "回中", sum(predicted == target[:-TEST_DATA_LENGTH]), "回で的中（", round(sum(predicted == target[:-TEST_DATA_LENGTH]) / len(target[:-TEST_DATA_LENGTH]) * 100), "%）")
```

* 現在の天気が1時間後の降雨の有無の判断に最も重要である
* 次に相対湿度を見ることで降雨を判定可能である。

***
## 30日分の降雨を予測する

```python
predicted = clf.predict(src[-TEST_DATA_LENGTH:])
```

```python
fig, ax1 = plt.subplots(figsize=(13, 5))
index = [datetime.datetime.strptime(i, '%Y/%m/%d %H:%M') for i in df.index[-TEST_DATA_LENGTH:]]
ax1.plot(index, target[-TEST_DATA_LENGTH:], color='r')
ax2 = ax1.twinx()
ax2.plot(index, df['降水量(mm)'].values[-TEST_DATA_LENGTH:])
plt.show()
```

<font color="blue">青線：降水量</font>  
<font color="red">赤線：1時間後の雨を予測した結果</font>

### 予測した結果

```python
print(TEST_DATA_LENGTH, "回中", sum(predicted == target[-TEST_DATA_LENGTH:]), "回で的中（", round(sum(predicted == target[-TEST_DATA_LENGTH:]) / len(target[-TEST_DATA_LENGTH:]) * 100), "%）")
```
